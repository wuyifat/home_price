```python
# allPair2 = allPair.loc[allPair['zipcode'] != "\\N" & allPair['yearbuilt'] != "\\N"]
allPair2 = allPair.loc[allPair['zipcode'] != "\\N"]
allPair2 = allPair2.loc[allPair2['yearbuilt'] != "\\N"]
```

```python
allPair2["zipcode"] = allPair2["zipcode"].astype(int)
allPair2["yearbuilt"] = allPair2["yearbuilt"].astype(int)
print allPair2.info()
```

```python
corr = allPair2.corr()
price_corr = corr["saleprice"]
price_corr = price_corr.to_dict()
# print price_corr
for ele in sorted(price_corr.items(), key = lambda x: -abs(x[1])):
    print("{0}: \t{1}".format(*ele))
```

```python
CA1 = allPair.loc[allPair["state"] == "CA"]
CA1 = CA1.loc[CA1['yearbuilt'] != "\\N"]
# allPair2 = allPair.loc[allPair['zipcode'] != "\\N"]
CA1["yearbuilt"] = CA1["yearbuilt"].astype(int)
# print CA1.info()
CA_zip = CA1.select_dtypes(include=['float64', 'int64'])
CA_zip["zipcode"] = CA1["zipcode"]
# print CA_zip.info()
CA_zipCoded = pd.get_dummies(CA_zip)
CA_zipCoded["saleprice"] = CA1["saleprice"]

corr_CA = CA_zipCoded.corr()
price_corr = corr_CA["saleprice"]
price_corr = price_corr.to_dict()
# print price_corr
for ele in sorted(price_corr.items(), key = lambda x: -abs(x[1])):
    print("{0}: \t{1}".format(*ele))
```
